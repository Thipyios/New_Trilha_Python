### __Exercício 2: Juntando os datasets__

1. Constrindo um novo ``DataFrame`` utilizando um ``MultiIndex``.

```python
# convertendo a coluna data para o tipo datetime
estacao_ssa_2003['DATA (YYYY-MM-DD)'] = pd.to_datetime(estacao_ssa_2003['DATA (YYYY-MM-DD)'])
estacao_ssa_2013['DATA (YYYY-MM-DD)'] = pd.to_datetime(estacao_ssa_2013['DATA (YYYY-MM-DD)'])
estacao_ssa_2023['Data'] = pd.to_datetime(estacao_ssa_2023['Data'])


# pegando listas com os anos, meses e dias de cada dataframe
anos = pd.concat([estacao_ssa_2003['DATA (YYYY-MM-DD)'].dt.year,
                  estacao_ssa_2013['DATA (YYYY-MM-DD)'].dt.year,
                  estacao_ssa_2023['Data'].dt.year])
anos = anos.values

meses = pd.concat([estacao_ssa_2003['DATA (YYYY-MM-DD)'].dt.month,
                   estacao_ssa_2013['DATA (YYYY-MM-DD)'].dt.month,
                   estacao_ssa_2023['Data'].dt.month])
meses = meses.values

dias = pd.concat([estacao_ssa_2003['DATA (YYYY-MM-DD)'].dt.day,
                  estacao_ssa_2013['DATA (YYYY-MM-DD)'].dt.day,
                  estacao_ssa_2023['Data'].dt.day])


# criando indice
index = pd.MultiIndex.from_arrays([anos, meses, dias], names=['ano', 'mes', 'dia'])
```

```python
# removendo colunas extraas
estacao_ssa_2003.drop(columns=['Unnamed: 19'], inplace=True)
estacao_ssa_2023.drop(columns=['Unnamed: 19'], inplace=True)
```

```python
# unificando os dataframes
print("Colunas antes:")
colunas_2003 = estacao_ssa_2003.columns.values.tolist()
colunas_2013 = estacao_ssa_2013.columns.values.tolist()
colunas_2023 = estacao_ssa_2023.columns.values.tolist()

print(colunas_2003)
print(colunas_2013)
print(colunas_2023)

# renomeando colunas diferentes
print("\nColunas diferentes:")
for c_2003, c_2013, c_2023 in zip(colunas_2003, colunas_2013, colunas_2023):
    if c_2003 != c_2013 or c_2003 != c_2023:
        print(f"'{c_2003}' | '{c_2013}' | '{c_2023}'")
        estacao_ssa_2013.rename(columns={c_2013: c_2003}, inplace=True)
        estacao_ssa_2023.rename(columns={c_2023: c_2003}, inplace=True)


# imprimindo colunas depois
print("\nColunas depois:")
colunas_2003 = estacao_ssa_2003.columns.values.tolist()
colunas_2013 = estacao_ssa_2013.columns.values.tolist()
colunas_2023 = estacao_ssa_2023.columns.values.tolist()

print(colunas_2003)
print(colunas_2013)
print(colunas_2023)

```

```python
# criando dataframa multiindexado por ano, mes e dia
estacao_ssa = pd.concat([estacao_ssa_2003, estacao_ssa_2013, estacao_ssa_2023]).drop(columns=['DATA (YYYY-MM-DD)', 'HORA (UTC)'])
estacao_ssa.set_index(index, inplace=True)
estacao_ssa
```

2. Construindo novo ``DataFrame`` utilizando ``MultiIndex`` como colunas contendo os parâmetros agrupados (precipitações, pressão, radiação, temperatura, umidade, velocidade do vento) 

```python
# criando multiindex para as colunas
nivel_1 = ['precipitacao'] + ['pressao']*3 + ['radiacao'] + ['temperatura']*6 + ['umidade']*3 + ['velocidade do vento']*3
nivel_2 = colunas_2003[2:]

# definindo multiindex para as colunas
index_col = pd.MultiIndex.from_arrays([nivel_1, nivel_2], names=['medicao', 'tipo'])

estacao_ssa.columns = index_col
estacao_ssa
```
